```python
sales_by_product = df.groupby("PRODUCTLINE")["SALES"].sum()
print(sales_by_product )
```

```python
sales_by_product =sales_by_product.sort_values(ascending=False)
sales_by_product
```

```python
plt.figure(figsize=(10,6))
plt.bar(sales_by_product.index , sales_by_product.values)
plt.title('Sales by Product Line')
plt.xlabel('Product line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
```

```python
sales_by_country = df.groupby('COUNTRY')['SALES'].sum()
sales_by_country = sales_by_country.sort_values(ascending=False)
print(sales_by_country)
```

```python
plt.figure(figsize=(12,6))
plt.bar(sales_by_country.index , sales_by_country.values)
plt.title('Sales by Country')
plt.xlabel('Countries')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
sales_by_year = df.groupby('YEAR_ID')['SALES'].sum()
sales_by_year = sales_by_year.sort_values(ascending=False)
sales_by_year
```

```python
plt.figure(figsize=(5 , 5))
plt.bar(sales_by_year.index , sales_by_year.values)
plt.xlabel('Sales By Year')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
```

```python
sales_by_month = df.groupby('MONTH_ID')['SALES'].sum()
sales_by_month = sales_by_month.sort_values(ascending=False)
sales_by_month
```

```python
top_customer = df.groupby('CUSTOMERNAME')['SALES'].sum()
top_customer = top_customer.sort_values(ascending=False)
top_customer.head(10)
```

```python
top_10 = (df.groupby('CUSTOMERNAME')['SALES'].sum()
         .sort_values(ascending=False)
         .head(10))
top_10
```

```python
plt.figure(figsize = (10, 6))
plt.bar(top_10.index ,top_10.values)

plt.title('Top 10 Customer')
plt.xlabel('Customer')
plt.ylabel('Total Sales')
plt.xticks(rotation = 90)
```
